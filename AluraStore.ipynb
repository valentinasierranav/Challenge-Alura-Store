# **Introducción al análisis comparativo**

El objetivo de este análisis es evaluar el desempeño de cuatro plataformas de venta (Tienda 1, Tienda 2, Tienda 3 y Tienda 4) con el fin de identificar cuál de ellas representa la mejor opción para continuar con la inversión del Sr. Juan. Para tomar una decisión informada, se realiza un análisis basado en datos que permita comparar el rendimiento de cada tienda desde diferentes perspectivas.

**Justificación del análisis**

Aunque todas las tiendas comercializan productos similares, su desempeño puede variar debido a factores como el volumen de ventas, los costos asociados y la percepción de los clientes. Por esta razón, no es suficiente analizar únicamente los ingresos generados por cada tienda. Un análisis integral permite comprender mejor el comportamiento de cada plataforma y reducir el riesgo al momento de tomar una decisión de inversión.

**Métricas clave del análisis**

Para realizar la comparación entre tiendas se analizarán varias métricas que permiten evaluar diferentes aspectos del negocio:

**Ingresos totales (Indicador financiero)**
Permite identificar qué tienda genera mayor volumen de ventas y cuál tiene mayor potencial de mercado.


**Calificación promedio de los clientes (Indicador de calidad)**
Refleja el nivel de satisfacción de los clientes con los productos o el servicio ofrecido en cada tienda.

**Costo promedio de envío (Indicador de eficiencia operativa)**
Ayuda a entender el impacto de los costos logísticos en cada tienda, lo cual puede afectar directamente la rentabilidad.

**Ventas por categoría y producto (Indicador de demanda)**
Permite identificar cuáles productos o categorías tienen mayor demanda dentro de cada tienda.



A partir del análisis de estas métricas será posible comparar el rendimiento de las cuatro tiendas y determinar cuál presenta mejores condiciones para una inversión sostenible a largo plazo.

# **Análisis de Facturación**

```python
# 1️⃣ Análisis de facturación
# En esta sección calculamos el ingreso total generado por cada tienda.
# Para ello, sumamos los valores de la columna 'Precio', que representa
# el valor de cada venta realizada.

# Lista con los DataFrames de las tiendas
lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]

# Usamos 'enumerate' para recorrer cada tienda e identificarla con un número
for i, df_tienda in enumerate(lista_de_tiendas, start=1):

    # Imprimimos el nombre de la tienda para identificar el resultado
    print(f'Tienda {i}')

    # Calculamos el ingreso total sumando los precios de todos los productos vendidos
    ingreso_total = df_tienda['Precio'].sum()

    # Mostramos el resultado con formato de moneda
    print(f'El ingreso total fue: ${ingreso_total:,.2f}')

    # Separador visual para facilitar la lectura
    print("-" * 40)
```

El análisis de facturación permite observar el ingreso total generado por cada tienda. Esta métrica proporciona una primera aproximación del rendimiento de cada plataforma y del volumen de ventas que manejan. Sin embargo, para tomar una decisión completa también es necesario considerar otros factores como los costos de envío, la satisfacción de los clientes y la demanda por categoría de producto.

```python
# 2️⃣ Ventas por categoría
# En esta sección analizamos cuántos productos se han vendido por cada categoría
# dentro de cada tienda. Esto permite identificar cuáles categorías tienen
# mayor demanda en cada plataforma.

# Lista con los DataFrames de las tiendas
lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]

# Recorremos cada tienda usando enumerate para identificarla con un número
for i, df_tienda in enumerate(lista_de_tiendas, start=1):

    # Imprimimos el nombre de la tienda
    print(f"Tienda {i} - Ventas por categoría")

    # Agrupamos los datos por categoría de producto
    # .size() cuenta cuántas ventas hay en cada categoría
    # reset_index() convierte el resultado en un DataFrame
    # sort_values() ordena las categorías de mayor a menor número de ventas
    resumen = (
        df_tienda
        .groupby('Categoria_producto')
        .size()
        .reset_index(name='Cantidad_vendida')
        .sort_values(by='Cantidad_vendida', ascending=False)
    )

    # Mostramos el resultado
    print(resumen)

    # Separador visual
    print("-" * 50)
```

```python
# Agregamos una columna que muestre el importe total vendido por categoría
# Esto permite ver no solo cuántas ventas hubo por categoría, sino también
# cuánto dinero generó cada una.

# Recorremos cada tienda
for i, df_tienda in enumerate(lista_de_tiendas, start=1):

    print(f'Tienda {i} - Ventas por categoría')

    # Agrupamos por categoría de producto
    # Calculamos:
    # - Cantidad_vendida: número de ventas por categoría
    # - Importe_total: suma total del precio de los productos vendidos
    resumen_ventas = (
        df_tienda
        .groupby('Categoria_producto')
        .agg(
            Cantidad_vendida=('Categoria_producto', 'size'),
            Importe_total=('Precio', 'sum')
        )
        .reset_index()
        .sort_values(by='Cantidad_vendida', ascending=False)
    )

    # Formateamos el importe total para que se vea como valor monetario
    resumen_ventas['Importe_total'] = resumen_ventas['Importe_total'].apply(lambda x: f"${x:,.2f}")

    # Mostramos la tabla
    print(resumen_ventas.to_string(index=False))

    # Separador visual
    print("-" * 50)
```

**Esto permite ver que:**

Una categoría puede vender muchas unidades pero generar menos dinero.

Otra puede vender menos pero tener precios más altos.


Esto permite identificar qué tipos de productos tienen mayor impacto en los ingresos de cada tienda.


En este caso, vemos que aunque "Muebles" es la categoria mas vendida, "electrónicos" y "electrodomésticos" son las que tienen un mayor importe de ventas

# **Calificación promedio de la tienda**

```python
# 3️⃣ Calificación promedio de la tienda
# En esta sección calculamos la calificación promedio de los clientes
# para cada tienda. Esto permite evaluar el nivel de satisfacción
# de los clientes con los productos o el servicio.

# Lista con los DataFrames de las tiendas
lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]

# Recorremos cada tienda usando enumerate
for i, df_tienda in enumerate(lista_de_tiendas, start=1):

    # Imprimimos el nombre de la tienda
    print(f'Tienda {i}')

    # Calculamos la calificación promedio de los clientes
    calificacion_promedio = df_tienda['Calificacion'].mean()

    # Mostramos el resultado con dos decimales
    print(f'La calificación promedio de la tienda fue: {calificacion_promedio:,.2f}')

    # Separador visual
    print("-" * 40)
```

La **calificación promedio** permite evaluar:

*  Qué tienda tiene clientes más satisfechos.

*  Si existe alguna tienda con peores valoraciones.

*  Qué plataforma podría tener mejor reputación entre los clientes.

Este resultado confirma los valores promedio de calificación que ya se habían observado durante el análisis exploratorio de los datos de cada tienda.

# **Productos más y menos vendidos**

```python
# 4️⃣ Productos más y menos vendidos
# En esta sección analizamos cuántas veces se vendió cada producto en cada tienda.
# Esto permite identificar cuáles productos tienen mayor demanda
# y cuáles presentan menor volumen de ventas.

# Lista con los DataFrames de las tiendas
lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]

# Recorremos cada tienda
for i, df_tienda in enumerate(lista_de_tiendas, start=1):

    # Imprimimos el nombre de la tienda
    print(f"Tienda {i} - Ventas por producto")

    # Agrupamos por producto
    # size() cuenta cuántas ventas hay por producto
    # reset_index() convierte el resultado en DataFrame
    # sort_values() ordena los productos de mayor a menor número de ventas
    resumen_productos = (
        df_tienda
        .groupby('Producto')
        .size()
        .reset_index(name='Cantidad_vendida')
        .sort_values(by='Cantidad_vendida', ascending=False)
    )

    # Mostramos la tabla completa
    print(resumen_productos.to_string(index=False))

    # Separador visual
    print("-" * 50)
```

Este análisis permite identificar cuáles productos tienen mayor y menor volumen de ventas en cada tienda. La información ayuda a comprender las preferencias de los clientes y a detectar qué artículos presentan mayor demanda dentro de cada plataforma.

# **Análisis adicional para identificar más claramente los productos destacados en ventas**

```python
# Aunque el listado completo permite ver la cantidad de ventas por producto,
# la tabla puede resultar extensa y poco práctica para identificar rápidamente
# los artículos más relevantes.

# Por esta razón, se implementa un análisis adicional que permite identificar
# de forma más clara:
# - el producto más vendido
# - el producto menos vendido
# - el Top 5 de productos más vendidos
# - el Top 5 de productos menos vendidos

# Esto facilita interpretar los resultados y destacar los productos
# con mayor y menor demanda en cada tienda.



# Análisis adicional para identificar más claramente los productos destacados en ventas

lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df_tienda in enumerate(lista_de_tiendas, start=1):
    print(f"\nTienda {i} - Productos más y menos vendidos")

    # Agrupar por producto y contar cuántas ventas tuvo cada uno
    ventas_por_producto = (
        df_tienda
        .groupby('Producto')
        .size()
        .reset_index(name='Cantidad_vendida')
        .sort_values(by='Cantidad_vendida', ascending=False)
    )

    # Producto más vendido
    top_producto = ventas_por_producto.iloc[0]

    # Producto menos vendido
    bottom_producto = ventas_por_producto.iloc[-1]

    print(f"Producto más vendido: {top_producto['Producto']} ({top_producto['Cantidad_vendida']} ventas)")
    print(f"Producto menos vendido: {bottom_producto['Producto']} ({bottom_producto['Cantidad_vendida']} ventas)")

    # Mostrar los 5 productos más vendidos
    print("\nTop 5 productos más vendidos:")
    print(ventas_por_producto.head(5).to_string(index=False))

    # Mostrar los 5 productos menos vendidos
    print("\nTop 5 productos menos vendidos:")
    print(ventas_por_producto.tail(5).to_string(index=False))

    print("-" * 60)
```

El código muestra para cada tienda:

🥇 Producto más vendido

📉 Producto menos vendido

🔝 Top 5 productos con más ventas

🔻 Top 5 productos con menos venta

# **Ahora analicemos si entre mas ventas hay realmente una facturación mayor**

Un producto puede vender mucho pero no necesariamente generar más ingresos si su precio es bajo.

Veamos:

📦 Cantidad de ventas por producto

💰 Ingresos generados por cada producto

🔝 Top productos que venden más

🔻 Productos con menor rotación

```python
# Análisis adicional: verificar si una mayor cantidad de ventas también implica
# una mayor facturación por producto.

lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df_tienda in enumerate(lista_de_tiendas, start=1):
    print(f"\nTienda {i} - Relación entre ventas y facturación por producto")

    # Agrupar por producto y calcular:
    # - Cantidad_vendida: número de veces que se vendió el producto
    # - Importe_total: suma del precio de todas las ventas de ese producto
    ventas_por_producto = (
        df_tienda
        .groupby('Producto')
        .agg(
            Cantidad_vendida=('Producto', 'size'),
            Importe_total=('Precio', 'sum')
        )
        .reset_index()
        .sort_values(by='Cantidad_vendida', ascending=False)
    )

    # Formatear el importe total para que se vea como valor monetario
    ventas_por_producto['Importe_total'] = ventas_por_producto['Importe_total'].apply(lambda x: f"${x:,.2f}")

    # Identificar el producto más vendido
    top_producto = ventas_por_producto.iloc[0]

    # Identificar el producto menos vendido
    bottom_producto = ventas_por_producto.iloc[-1]

    print(f"\nProducto más vendido: {top_producto['Producto']} ({top_producto['Cantidad_vendida']} ventas, {top_producto['Importe_total']})")
    print(f"Producto menos vendido: {bottom_producto['Producto']} ({bottom_producto['Cantidad_vendida']} ventas, {bottom_producto['Importe_total']})")

    # Mostrar los 5 productos más vendidos
    print("\nTop 5 productos más vendidos:")
    print(ventas_por_producto.head(5).to_string(index=False))

    # Mostrar los 5 productos menos vendidos
    print("\nTop 5 productos menos vendidos:")
    print(ventas_por_producto.tail(5).to_string(index=False))

    print("-" * 60)
```

A partir de este análisis se puede observar que un mayor número de ventas no necesariamente implica una mayor facturación. Esto se debe a que algunos productos tienen precios más altos que otros, por lo que incluso con menos ventas pueden generar ingresos significativos. Además, este comportamiento sugiere que cada tienda puede estar enfocada en distintos tipos de productos o en nichos específicos de mercado.
